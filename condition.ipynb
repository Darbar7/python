```python
#11.Three students — Aman, Ravi, and Sneha — scored 56, 74, and 68 marks respectively.
 #Write a program to print who got the highest marks.
 #Output:Ravi got the highest marks.

Aman = 56
Ravi = 74
Sneha = 68
highest = max(Aman,Ravi,Sneha)
if highest == Aman: 
    print('Aman got the highest marks')
elif highest == Ravi:
    print('Ravi got the highest marks')
else:
    print('Sneha got the highest marks')
```
