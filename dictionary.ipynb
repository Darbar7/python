```python
#Add multiple items to a dictionary and use .popitem() twice. Observe the removed items.
tools = {'git': 'vcs'}
new_tools = {'docker': 'container', 'k8s': 'orchestration', 'jenkins': 'ci-cd'}
tools.popitem(),new_tools.popitem()
```

```python
#Merge two dictionaries using .update().
cloud_skills = {'AWS': 'EC2', 'Docker': 'Containers'}
devops_tools = {'Kubernetes': 'Orchestration', 'Jenkins': 'CI/CD', 'Terraform': 'IaC'}
cloud_skills.update(devops_tools)
print(cloud_skills)
```

```python
#Use .update() to change the value of an existing key in the dictionary.
skills = {'Python': 'Beginner', 'AWS': 'Intermediate'}
updates = {'Python': 'Advanced', 'Docker': 'Beginner'}
skills.update(updates)
print("updates_skills",skills)

```
